## Loading the necessary libraries
Load the necessary libraries to analyse and visualize the dataset

Load the dataset into pandas DataFrame and print top 5 rows

 - **Year:** Quantitative variable (integer)
 - **Life expectancy at birth (years):** Quantitative variable (float)
 - **GDP:** Quantitative variable

### Distribution of Life Expectancy

```python
# Plot the distribution of life expectancy
plt.figure(figsize=(10, 6))
sns.histplot(df['Life expectancy at birth (years)'])
plt.title('Distribution of Life Expectancy at Birth')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Frequency')
plt.show()
```

The histogram shows that the life expectancy variable's distribution is skewed towards the left

```python
sns.boxplot(df['Life expectancy at birth (years)'])
plt.title('Life expectancy at birth in years distribution')
plt.xlabel('Life expectancy at birth in years')
plt.show()
```

By visualizing the boxplot we can notice that are many outliers in our variable

### Distribution of GDP

```python
# Plot the distribution of GDP
plt.figure(figsize=(10, 6))
sns.histplot(df['GDP'], kde=True)
plt.title('Distribution of GDP')
plt.xlabel('GDP')
plt.ylabel('Frequency')
plt.show()
```

### Life Expectancy Over Time
We'll plot life expectancy over time for each country to see trends.

```python
# Plot life expectancy over time for each country
plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x='Year', y='Life expectancy at birth (years)', hue='Country')
plt.title('Life Expectancy at Birth Over Time')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country')
plt.show()
```

### GDP Over Time
We'll plot GDP over time for each country to see economic trends.

```python
# Plot GDP over time for each country
plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
plt.title('GDP Over Time')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Country')
plt.show()
```

### Relationship Between Life Expectancy and GDP
We'll explore the relationship between life expectancy and GDP.

```python
# Plot the relationship between life expectancy and GDP
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GDP', y='Life expectancy at birth (years)', hue='Country')
plt.title('Relationship Between Life Expectancy and GDP')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country')
plt.show()
```

### Logarithmic Transformation
Because the values for the GDP are so high we are not able to visualize the relationship. Let's address this by transforming the GDP to more manageable scale. We'll use a logarithmic transformation, which is a common technique to handle large ranges of values and make trends more visible.

```python
# Apply logarithmic transformation to GDP 
df['Log GDP'] = np.log(df['GDP'])
```

```python
# Plot log GDP over time for each country
plt.figure(figsize=(14, 8))
sns.lineplot(data=df, x='Year', y='Log GDP', hue='Country')
plt.title('Log GDP Over Time')
plt.xlabel('Year')
plt.ylabel('Log GDP')
plt.legend(title='Country')
plt.show()
```

```python
# Plot the relationship between life expectancy and log GDP
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Log GDP', y='Life expectancy at birth (years)', hue='Country')
plt.title('Relationship Between Life Expectancy and Log GDP')
plt.xlabel('Log GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country')
plt.show()
```
